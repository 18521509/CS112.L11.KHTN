```python
#Đọc dữ liệu, loại bỏ cột NaN

path = '/content/drive/MyDrive/Matrix Profile/data/data.csv'
df = pd.read_csv(path)
```

```python
data_size = 100000
df.fillna(0)
x = np.asarray(df)
y = x.T
data = y[3:]
data = data[:,:data_size].astype(float)
```

```python
A,B = data[0],data[1]
m = 200
res = stump(A,m,B,ignore_trivial = False)
```

```python
mf,mf_ind = res[0],res[1]
```

```python
A_ind = res[:,0].argmin()

B_ind = res[A_ind,1]
```
