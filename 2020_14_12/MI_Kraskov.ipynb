#Plot

```
import matplotlib.pyplot as plt
import os

dirX = os.path.join(dir,'x_layer')
dirY = os.path.join(dir,'y_layer')

miX = [None]*5
miY = [None]*5

for layer in range(5):
    fileX = os.path.join(dirX,f'X_{layer:0>2}.txt')
    fileY = os.path.join(dirY,f'Y_{layer:0>2}.txt')
    dataX = list(map(float,open(fileX,'r').read().split()))
    dataY = list(map(float,open(fileY,'r').read().split()))
    miX[layer] = dataX.copy()
    miY[layer] = dataY.copy()

def plot(data,name):
    plt.figure(figsize=(16,9))
    plt.title('Mutual information: '+ name)
    for layer in range(len(data)):
        plt.plot(data[layer],label = f'layer {layer}')
    plt.legend(loc = 'upper right')
    plt.show()
plot(miX,'X - Layer')
plot(miY,'Y - Layer')
```
